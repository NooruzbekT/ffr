```python
import pandas as pd
import plotly.express as px



df['predicted_language'] = df['predicted_language'].str.strip().str.capitalize()  # Уберём пробелы, нормализуем

# Подсчёт количества
lang_counts = df['predicted_language'].value_counts().reset_index()
lang_counts.columns = ['Language', 'Count']

# Вычисление процентов
total = lang_counts['Count'].sum()
lang_counts['Percent'] = round(lang_counts['Count'] / total * 100, 2)

# 🌈 Интерeктивная круговая диаграмма
fig = px.pie(
    lang_counts,
    values='Count',
    names='Language',
    title='Language Distribution in Dataset',
    color_discrete_sequence=px.colors.qualitative.Bold,
    hole=0.4
)

fig.update_traces(textposition='inside', textinfo='percent+label', pull=[0.05]*len(lang_counts))


fig.update_layout(
    showlegend=True,
    font=dict(size=16),
    title_font_size=24
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Пример: df['date'] содержит значения типа 20250105
df['base_ymd'] = pd.to_datetime(df['base_ymd'].astype(str), format='%Y%m%d')
# Удалим декабрь 2024

# Создаём колонку "месяц-год" в виде "2025-01"
df['month'] = df['base_ymd'].dt.to_period('M').astype(str)
df = df[df['month'] != '2024-12']
# Группировка по месяцам
month_counts = df['month'].value_counts().sort_index().reset_index()
month_counts.columns = ['Month', 'Count']

# Построим интерактивный график
fig = px.bar(
    month_counts,
    x='Month',
    y='Count',
    title='📅 Количество комментариев по месяцам',
    labels={'Month': 'Месяц', 'Count': 'Количество комментариев'},
    color='Count',
    color_continuous_scale='Turbo'
)

fig.update_layout(
    xaxis_tickangle=-45,
    font=dict(size=14),
    title_font_size=22
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px




# Группировка: количество по месяцу и языку
monthly_lang_counts = df.groupby(['month', 'predicted_language']).size().reset_index(name='count')

# Вычислим общее кол-во за каждый месяц
total_per_month = df.groupby('month').size().reset_index(name='total')

# Объединяем
merged = monthly_lang_counts.merge(total_per_month, on='month')
merged['percent'] = round(merged['count'] / merged['total'] * 100, 2)

# Сортировка по дате
merged = merged.sort_values('month')

# 🌈 Построим stacked bar (кол-во)
fig1 = px.bar(
    merged,
    x='month',
    y='count',
    color='predicted_language',
    title='📊 Количество обращений по языкам (по месяцам)',
    labels={'count': 'Количество', 'month': 'Месяц'},
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig1.update_layout(
    barmode='stack',
    xaxis_tickangle=-45,
    font=dict(size=14),
    title_font_size=22
)
fig1.show()

# 🌈 Теперь график с % (normalized stacked bar)
fig2 = px.bar(
    merged,
    x='month',
    y='percent',
    color='predicted_language',
    title='📊 Процент языков по месяцам',
    labels={'percent': '%', 'month': 'Месяц'},
    color_discrete_sequence=px.colors.qualitative.Set3
)

fig2.update_layout(
    barmode='stack',
    xaxis_tickangle=-45,
    yaxis=dict(title='Процент', ticksuffix='%'),
    font=dict(size=14),
    title_font_size=22
)
fig2.show()
```

```python
df['sentiment'] = df['sentiment'].str.strip().str.capitalize()  # Normalize

# Подсчёт
sentiment_counts = df['sentiment'].value_counts().reset_index()
sentiment_counts.columns = ['Sentiment', 'Count']
sentiment_counts['Percent'] = round(sentiment_counts['Count'] / sentiment_counts['Count'].sum() * 100, 2)

# Pie chart
import plotly.express as px

fig = px.pie(
    sentiment_counts,
    names='Sentiment',
    values='Count',
    title='🎭 Общая тональность обращений',
    color_discrete_sequence=px.colors.qualitative.Set1,
    hole=0.4
)
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
# Кол-во по месяцам и тональности
monthly_sentiment = df.groupby(['month', 'sentiment']).size().reset_index(name='count')

# Общее число обращений в месяц
monthly_total = df.groupby('month').size().reset_index(name='total')

# Объединяем
merged = monthly_sentiment.merge(monthly_total, on='month')
merged['percent'] = round(merged['count'] / merged['total'] * 100, 2)

# Сортировка
merged = merged.sort_values('month')

# 1️⃣ Stacked bar: кол-во по тональности
fig1 = px.bar(
    merged,
    x='month',
    y='count',
    color='sentiment',
    title='🗓️ Количество обращений по тональности (по месяцам)',
    labels={'count': 'Количество', 'month': 'Месяц'},
    color_discrete_sequence=px.colors.qualitative.Pastel
)
fig1.update_layout(
    barmode='stack',
    xaxis_tickangle=-45,
    font=dict(size=14),
    title_font_size=22
)
fig1.show()

# 2️⃣ Normalized stacked bar (в процентах)
fig2 = px.bar(
    merged,
    x='month',
    y='percent',
    color='sentiment',
    title='🧠 Процент тональностей по месяцам',
    labels={'percent': '%', 'month': 'Месяц'},
    color_discrete_sequence=px.colors.qualitative.Set3
)
fig2.update_layout(
    barmode='stack',
    xaxis_tickangle=-45,
    yaxis=dict(title='Процент', ticksuffix='%'),
    font=dict(size=14),
    title_font_size=22
)
fig2.show()
```

```python
import plotly.express as px



# Фильтруем негативные
neg = df[df['sentiment'] == 'Negative']
neg_lang = neg['predicted_language'].value_counts().reset_index()
neg_lang.columns = ['Language', 'Count']
neg_lang['Percent'] = round(neg_lang['Count'] / neg_lang['Count'].sum() * 100, 2)

# Pie chart для негативных
fig1 = px.pie(
    neg_lang,
    names='Language',
    values='Count',
    title='🔴 Языковое распределение негативных отзывов',
    color_discrete_sequence=px.colors.qualitative.Set1,
    hole=0.4
)
fig1.update_traces(textinfo='percent+label')
fig1.show()

# Фильтруем позитивные
pos = df[df['sentiment'] == 'Positive']
pos_lang = pos['predicted_language'].value_counts().reset_index()
pos_lang.columns = ['Language', 'Count']
pos_lang['Percent'] = round(pos_lang['Count'] / pos_lang['Count'].sum() * 100, 2)

# Pie chart для позитивных
fig2 = px.pie(
    pos_lang,
    names='Language',
    values='Count',
    title='🟢 Языковое распределение позитивных отзывов',
    color_discrete_sequence=px.colors.qualitative.Set2,
    hole=0.4
)
fig2.update_traces(textinfo='percent+label')
fig2.show()
```

```python
import plotly.express as px



# Функция для построения pie-графика
def plot_lang_pie_by_type(client_type, title_icon=''):
    subset = df[df['typeclient'] == client_type]
    lang_counts = subset['predicted_language'].value_counts().reset_index()
    lang_counts.columns = ['Language', 'Count']
    lang_counts['Percent'] = round(lang_counts['Count'] / lang_counts['Count'].sum() * 100, 2)

    fig = px.pie(
        lang_counts,
        names='Language',
        values='Count',
        title=f'{title_icon} Языковое распределение обращений: {client_type}',
        color_discrete_sequence=px.colors.qualitative.Set3,
        hole=0.4
    )
    fig.update_traces(textinfo='percent+label')
    fig.show()

# 🔵 ФЛ
plot_lang_pie_by_type('Фл', '👤')

# 🟢 ЮЛ
plot_lang_pie_by_type('Юл', '🏢')

# 🟡 ИПЧ
plot_lang_pie_by_type('Ипч', '📦')
```

```python
import pandas as pd
import plotly.express as px

# 🔹 Категориальные колонки
category_cols = ['forced_services', 'slow_service_long_wait', 'app_banking_issues', 'open_close_issues', 'false_information', 'staff_incompetence',
                 'transfer_payment_issues', 'ui_display_issues', 'no_notifications', 'tariffs_limits_issues', 'account_statement_issues', 'topup_withdrawal_issues',
                 'tech_support_issues', 'credit_issues', 'repayment_issues', 'other', 'atm_ipt_issues']

import plotly.express as px
import pandas as pd

# Предыдущие шаги оставляем как есть:
df_binary = df[category_cols].copy()
category_counts = df_binary.sum().reset_index()
category_counts.columns = ['Category', 'Count']

# Добавим "Без категории"
all_zero = (df_binary.sum(axis=1) == 0).sum()
category_counts.loc[len(category_counts)] = ['Без категории', all_zero]

# Подсчёт общего количества строк
total = len(df_binary)
category_counts['Percent'] = round(category_counts['Count'] / total * 100, 2)


final_df = category_counts.sort_values('Percent', ascending=False).reset_index(drop=True)

# 🎨 Donut chart с hover-only
fig = px.pie(
    final_df,
    names='Category',
    values='Count',
    title='📊 Распределение текстов по категориям',
    color_discrete_sequence=px.colors.qualitative.Pastel,
    hole=0.45,
    hover_data=['Percent']  # показываем процент при наведении
)


fig.update_traces(textinfo='none')

# 🛠 Улучшаем читаемость и вынос легенды
fig.update_layout(
    title_font_size=20,
    font=dict(size=14),
    legend_title_text='Категория',
    showlegend=True,
    height=700,
    width=1000,
    margin=dict(l=50, r=250, t=80, b=50),
    legend=dict(
        x=1.05,
        y=1,
        traceorder='normal',
        font=dict(size=12)
    )
)

fig.show()

```
