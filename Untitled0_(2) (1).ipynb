```python
label_list = train.features['ner_tags'].feature.names
label2id = {l: i for i, l in enumerate(label_list)}
id2label = {i: l for i, l in enumerate(label_list)}

print(label2id)
```

```python
from transformers import AutoTokenizer

model_ckpt = "xlm-roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)

def tokenize_and_align_labels(example):
    tokenized_inputs = tokenizer(
        example["tokens"],
        truncation=True,
        padding="max_length",
        max_length=128,
        is_split_into_words=True
    )

    word_ids = tokenized_inputs.word_ids()
    previous_word_idx = None
    label_ids = []

    for word_idx in word_ids:
        if word_idx is None:
            label_ids.append(-100)
        elif word_idx != previous_word_idx:
            label_ids.append(example["ner_tags"][word_idx])  # первый сабтокен
        else:
            label_ids.append(-100)  # все остальные сабтокены игнорируются
        previous_word_idx = word_idx

    tokenized_inputs["labels"] = label_ids
    return tokenized_inputs


train_tok = train.map(tokenize_and_align_labels, batched=False)
valid_tok = valid.map(tokenize_and_align_labels, batched=False)
```

```python
from transformers import AutoModelForTokenClassification

model = AutoModelForTokenClassification.from_pretrained(
    model_ckpt,
    num_labels=len(label_list),
    id2label=id2label,
    label2id=label2id
)

model.gradient_checkpointing_enable()
```

```python
from transformers import TrainingArguments, Trainer
from seqeval.metrics import accuracy_score, f1_score, classification_report

def compute_metrics(p):
    predictions, labels = p
    predictions = predictions.argmax(axis=-1)

    true_labels = [
        [label_list[l] for l in label if l != -100]
        for label in labels
    ]
    true_preds = [
        [label_list[p] for (p, l) in zip(pred, label) if l != -100]
        for pred, label in zip(predictions, labels)
    ]
    return {
        "f1": f1_score(true_labels, true_preds),
        "accuracy": accuracy_score(true_labels, true_preds)
    }

training_args = TrainingArguments(
    output_dir="./xlmr-base-ner",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,       # безопасно для T4
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=2,
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    fp16=True,                           # ускоряет на T4
    logging_dir="./logs",
    logging_steps=50
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_tok,
    eval_dataset=valid_tok,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
test_tok = test.map(tokenize_and_align_labels, batched=False)
results = trainer.evaluate(test_tok)
print(results)
```

```python
from seqeval.metrics import classification_report

predictions, labels, _ = trainer.predict(test_tok)
preds = predictions.argmax(axis=-1)

true_labels = [
    [label_list[l] for l in label if l != -100]
    for label in labels
]
true_preds = [
    [label_list[p] for (p, l) in zip(pred, label) if l != -100]
    for pred, label in zip(preds, labels)
]

print(classification_report(true_labels, true_preds))
```

```python
import torch
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForTokenClassification
from tqdm import tqdm
from seqeval.metrics import classification_report, f1_score, accuracy_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Teacher
teacher_model = AutoModelForTokenClassification.from_pretrained(
    "/content/drive/MyDrive/BCC/Методы оптимизации /xlmr-base-ner-teacher"
)
teacher_model.to(device).eval()

# Student
student_model = AutoModelForTokenClassification.from_pretrained(
    "yasminsur/distilbert-kazakh-ner-2",
    num_labels=teacher_model.config.num_labels
)
student_model.to(device).train()

# Tokenizers
teacher_tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")
student_tokenizer = AutoTokenizer.from_pretrained("yasminsur/distilbert-kazakh-ner-2")
```

```python
def tokenize_teacher(example):
    tok = teacher_tokenizer(
        example["tokens"],
        truncation=True,
        padding="max_length",
        max_length=128,
        is_split_into_words=True
    )
    word_ids = tok.word_ids()
    labels = []
    prev = None
    for word_idx in word_ids:
        if word_idx is None:
            labels.append(-100)
        elif word_idx != prev:
            labels.append(example["ner_tags"][word_idx])
        else:
            labels.append(-100)
        prev = word_idx
    tok["labels"] = labels
    return tok

def tokenize_student(example):
    tok = student_tokenizer(
        example["tokens"],
        truncation=True,
        padding="max_length",
        max_length=128,
        is_split_into_words=True
    )
    word_ids = tok.word_ids()
    labels = []
    prev = None
    for word_idx in word_ids:
        if word_idx is None:
            labels.append(-100)
        elif word_idx != prev:
            labels.append(example["ner_tags"][word_idx])
        else:
            labels.append(-100)
        prev = word_idx
    tok["labels"] = labels
    return tok
```

```python
train_teacher = train.map(tokenize_teacher)
train_student = train.map(tokenize_student)
valid_student = valid.map(tokenize_student)

class NERDistillDataset(Dataset):
    def __init__(self, teacher_ds, student_ds):
        self.teacher_ds = teacher_ds
        self.student_ds = student_ds

    def __len__(self):
        return len(self.teacher_ds)

    def __getitem__(self, idx):
        t = self.teacher_ds[idx]
        s = self.student_ds[idx]
        return {
            "teacher_input_ids": torch.tensor(t["input_ids"]),
            "teacher_attention_mask": torch.tensor(t["attention_mask"]),
            "student_input_ids": torch.tensor(s["input_ids"]),
            "student_attention_mask": torch.tensor(s["attention_mask"]),
            "labels": torch.tensor(s["labels"]),
        }

train_dataset = NERDistillDataset(train_teacher, train_student)

# Валидатор можно оставить как есть
class SimpleNERDataset(Dataset):
    def __init__(self, ds):
        self.ds = ds
    def __len__(self): return len(self.ds)
    def __getitem__(self, idx):
        item = self.ds[idx]
        return {
            "input_ids": torch.tensor(item["input_ids"]),
            "attention_mask": torch.tensor(item["attention_mask"]),
            "labels": torch.tensor(item["labels"]),
        }

valid_dataset = SimpleNERDataset(valid_student)
```

```python
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=8)
```

```python
optimizer = torch.optim.AdamW(student_model.parameters(), lr=3e-5)
ce_loss_fn = torch.nn.CrossEntropyLoss(ignore_index=-100)
temperature = 3.0
alpha = 0.8
label_list = train.features["ner_tags"].feature.names
```

```python
def evaluate(model, loader):
    model.eval()
    all_preds, all_labels = [], []

    for batch in loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        with torch.no_grad():
            logits = model(input_ids=input_ids, attention_mask=attention_mask).logits
        preds = torch.argmax(logits, dim=-1)

        for pred, label in zip(preds, labels):
            true, pred_ = [], []
            for p, l in zip(pred, label):
                if l != -100:
                    true.append(label_list[l])
                    pred_.append(label_list[p])
            all_labels.append(true)
            all_preds.append(pred_)

    print(classification_report(all_labels, all_preds))
    return f1_score(all_labels, all_preds), accuracy_score(all_labels, all_preds)
```

```python
for epoch in range(5):
    student_model.train()
    total_loss = 0.0

    for batch in tqdm(train_loader):
        t_input_ids = batch["teacher_input_ids"].to(device)
        t_attn = batch["teacher_attention_mask"].to(device)
        s_input_ids = batch["student_input_ids"].to(device)
        s_attn = batch["student_attention_mask"].to(device)
        labels = batch["labels"].to(device)

        with torch.no_grad():
            t_logits = teacher_model(input_ids=t_input_ids, attention_mask=t_attn).logits

        s_logits = student_model(input_ids=s_input_ids, attention_mask=s_attn).logits

        # match shape
        t_logits = t_logits[:, :s_logits.size(1), :]

        # reshape
        s_flat = s_logits.view(-1, s_logits.size(-1))
        t_flat = t_logits.view(-1, t_logits.size(-1))
        lbl = labels.view(-1)

        loss_kd = F.kl_div(
            F.log_softmax(s_flat / temperature, dim=-1),
            F.softmax(t_flat / temperature, dim=-1),
            reduction="batchmean"
        ) * (temperature ** 2)

        loss_ce = ce_loss_fn(s_flat, lbl)
        loss = alpha * loss_ce + (1 - alpha) * loss_kd

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        total_loss += loss.item()

    print(f"\nEpoch {epoch+1} — Loss: {total_loss:.4f}")
    f1, acc = evaluate(student_model, valid_loader)
    print(f"F1: {f1:.4f} | Acc: {acc:.4f}\n")
```
