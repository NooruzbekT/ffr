```python
def get_xlm_embedding(texts, batch_size=16):
    embeddings = []
    with torch.no_grad():
        for i in tqdm(range(0, len(texts), batch_size)):
            batch = texts[i:i+batch_size]

            encoded = tokenizer(batch, padding=True, truncation=True, return_tensors="pt", max_length=512)
            input_ids = encoded['input_ids'].cuda()
            attention_mask = encoded['attention_mask'].cuda()

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            last_hidden_state = outputs.last_hidden_state


            input_mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size())
            sum_embeddings = torch.sum(last_hidden_state * input_mask_expanded, dim=1)
            sum_mask = input_mask_expanded.sum(dim=1)
            mean_pooled = sum_embeddings / sum_mask.clamp(min=1e-9)

            embeddings.append(mean_pooled.cpu())

    return torch.cat(embeddings, dim=0)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import AutoModel

class MultiLevelClassifier(nn.Module):
    def __init__(self, model_name="FacebookAI/xlm-roberta-base", projection_dim=256, num_classes=4,class_weights=None, kappa=0.2, freeze_encoder=False):
        super().__init__()
        self.kappa = kappa
        self.encoder = AutoModel.from_pretrained(model_name)
        self.class_weights = class_weights

        if freeze_encoder:
            for param in self.encoder.parameters():
                param.requires_grad = False

        self.hidden_size = self.encoder.config.hidden_size

        self.projector = nn.Sequential(
            nn.Linear(self.hidden_size, projection_dim),
            nn.ReLU(),
            nn.Linear(projection_dim, projection_dim)
        )

        self.classifier = nn.Linear(self.hidden_size, num_classes)

    def mean_pooling(self, last_hidden_state, attention_mask):
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size())
        sum_embeddings = torch.sum(last_hidden_state * input_mask_expanded, dim=1)
        sum_mask = input_mask_expanded.sum(dim=1)
        return sum_embeddings / sum_mask.clamp(min=1e-9)

    def nt_xent_loss(self, z, labels, temperature=0.05):
        sim = torch.matmul(z, z.T) / temperature
        labels = labels.unsqueeze(1)
        mask = torch.eq(labels, labels.T).float().to(z.device)
        exp_sim = torch.exp(sim) * (1 - torch.eye(len(z)).to(z.device))
        pos_sim = exp_sim * mask
        loss = -torch.log((pos_sim.sum(1) + 1e-9) / (exp_sim.sum(1) + 1e-9))
        return loss.mean()

    def forward(self, input_ids, attention_mask, labels=None):
        out = self.encoder(input_ids=input_ids, attention_mask=attention_mask).last_hidden_state
        pooled = self.mean_pooling(out, attention_mask)

        logits = self.classifier(pooled)
        z = F.normalize(self.projector(pooled), dim=1)

        output = {"logits": logits}

        if labels is not None:
            ce_loss = F.cross_entropy(logits, labels, weight=self.class_weights)
            cl_loss = self.nt_xent_loss(z, labels)
            total_loss = ce_loss + self.kappa * cl_loss
            output["loss"] = total_loss

        return output
```

```python
from torch.utils.data import Dataset
from sklearn.preprocessing import LabelEncoder
import random
import torch

class TextContrastiveDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, label_encoder=None, max_length=128):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_length = max_length

        if label_encoder is None:
            self.label_encoder = LabelEncoder()
            self.encoded_labels = self.label_encoder.fit_transform(labels)
        else:
            self.label_encoder = label_encoder
            self.encoded_labels = self.label_encoder.transform(labels)

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return {
            "text": self.texts[idx],
            "label": self.encoded_labels[idx]
        }

def simple_collate_fn(batch, tokenizer, max_length=128):
    texts = [x["text"] for x in batch]
    labels = torch.tensor([x["label"] for x in batch], dtype=torch.long)

    encoded = tokenizer(texts, padding=True, truncation=True, max_length=max_length, return_tensors="pt")

    return {
        "input_ids": encoded["input_ids"],
        "attention_mask": encoded["attention_mask"],
        "labels": labels
    }
```

```python
import torch
from sklearn.metrics import accuracy_score, f1_score
from tqdm import tqdm

def train(model, dataloader, optimizer, scheduler, device):
    model.train()
    total_loss = 0
    all_preds = []
    all_labels = []

    for batch in tqdm(dataloader):
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(
            input_ids=batch['input_ids'],
            attention_mask=batch['attention_mask'],
            labels=batch['labels']
        )
        loss = outputs["loss"]
        loss.backward()
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()

        total_loss += loss.item()
        preds = torch.argmax(outputs["logits"], dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(batch["labels"].cpu().numpy())

    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='macro')
    return total_loss / len(dataloader), acc, f1



def evaluate(model, dataloader, device):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(dataloader):
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(
                input_ids=batch['input_ids'],
                attention_mask=batch['attention_mask'],
                labels=batch['labels']
            )
            loss = outputs["loss"]
            total_loss += loss.item()
            preds = torch.argmax(outputs["logits"], dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(batch["labels"].cpu().numpy())

    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='macro')
    return total_loss / len(dataloader), acc, f1

```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def evaluate_with_report(model, dataloader, label_encoder, device='cuda'):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Eval"):
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(
                input_ids=batch['input_ids'],
                attention_mask=batch['attention_mask'],
                labels=batch['labels']
            )
            preds = torch.argmax(outputs["logits"], dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(batch["labels"].cpu().numpy())

    # === F1 / Precision / Recall
    class_names = label_encoder.classes_
    print("📋 Classification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))

    # === Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds, labels=np.arange(len(class_names)))
    plt.figure(figsize=(max(12, len(class_names) * 0.4), max(10, len(class_names) * 0.4)))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    return all_preds, all_labels
```

```python
from sklearn.manifold import TSNE

def visualize_embeddings(model, dataloader, label_encoder, device='cuda'):
    model.eval()
    all_embeds = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Embedding Extract"):
            batch = {k: v.to(device) for k, v in batch.items()}
            out = model.encoder(input_ids=batch["input_ids"], attention_mask=batch["attention_mask"])
            pooled = model.mean_pooling(out.last_hidden_state, batch["attention_mask"])
            all_embeds.append(pooled.cpu())
            all_labels.extend(batch["labels"].cpu().numpy())

    all_embeds = torch.cat(all_embeds, dim=0).numpy()
    all_labels = np.array(all_labels)

    tsne = TSNE(n_components=2, perplexity=30, random_state=42)
    reduced = tsne.fit_transform(all_embeds)

    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(reduced[:, 0], reduced[:, 1], c=all_labels, cmap='tab20', alpha=0.7)
    plt.colorbar(scatter, ticks=range(len(label_encoder.classes_)))
    plt.title("t-SNE Projection of Text Embeddings")
    plt.grid(True)
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import DataLoader
from transformers import AutoTokenizer, get_scheduler
import torch
from sklearn.utils.class_weight import compute_class_weight

# 1. Tokenizer для XLM-R (без "query:" префикса)
tokenizer = AutoTokenizer.from_pretrained("FacebookAI/xlm-roberta-base")

# 2. LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(df['combined'])
num_classes = len(label_encoder.classes_)
all_labels = label_encoder.transform(df['combined'])

# Считаем веса
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(all_labels),
    y=all_labels
)

class_weights = torch.tensor(class_weights, dtype=torch.float).to('cuda')

# 3. Разделение
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df['text_clean'], df['combined'], test_size=0.2, stratify=df['combined']
)

# 4. Dataset (предполагается, что TextContrastiveDataset уже без "query: ")
train_dataset = TextContrastiveDataset(train_texts.tolist(), train_labels.tolist(), tokenizer, label_encoder=label_encoder)
val_dataset = TextContrastiveDataset(val_texts.tolist(), val_labels.tolist(), tokenizer, label_encoder=label_encoder)

# 5. DataLoader
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True,
                          collate_fn=lambda x: simple_collate_fn(x, tokenizer))
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False,
                        collate_fn=lambda x: simple_collate_fn(x, tokenizer))

# 6. Модель
model = MultiLevelClassifier(
    model_name="FacebookAI/xlm-roberta-base",
    num_classes=num_classes, class_weights=class_weights
).cuda()

# 7. Optimizer и Scheduler
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
num_epochs = 8
num_training_steps = len(train_loader) * num_epochs

scheduler = get_scheduler(
    name="linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps
)

# 8. Train loop
best_f1 = 0.0
save_path = "/content/drive/MyDrive/BCC/42 тыс/best_model.pt"

for epoch in range(num_epochs):
    print(f"\n🔁 Epoch {epoch+1} / {num_epochs}")
    print("-" * 40)

    train_loss, train_acc, train_f1 = train(model, train_loader, optimizer, scheduler, device='cuda')
    print(f"✅ Train | Loss: {train_loss:.4f}, Acc: {train_acc:.4f}, F1: {train_f1:.4f}")

    val_loss, val_acc, val_f1 = evaluate(model, val_loader, device='cuda')
    print(f"📊 Val   | Loss: {val_loss:.4f}, Acc: {val_acc:.4f}, F1: {val_f1:.4f}")

    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), save_path)
        print(f"💾 Сохранили новую лучшую модель в {save_path} с F1 = {best_f1:.4f}")
    else:
        print(f"📉 F1 не улучшился, модель не сохраняется (текущий best = {best_f1:.4f})")

# 9. Отчёт и визуализация
evaluate_with_report(model, val_loader, label_encoder, device='cuda')
visualize_embeddings(model, val_loader, label_encoder, device='cuda')
```

```python
import torch
import torch.nn as nn
import onnx
import onnxruntime as ort
from onnxruntime.quantization import quantize_dynamic, QuantType
import numpy as np
import copy
from sklearn.metrics import accuracy_score, f1_score, classification_report
import time
import os
from tqdm import tqdm
import json
import pickle
```

```python
def measure_model_size(file_path, model_name="model"):
    """Измерить размер файла модели"""
    if os.path.exists(file_path):
        size_mb = os.path.getsize(file_path) / 1024 / 1024
        print(f"📏 {model_name} size: {size_mb:.2f} MB")
        return size_mb
    else:
        print(f"❌ File not found: {file_path}")
        return 0

def export_to_onnx(model, tokenizer, save_path, sample_input_text="Sample text for tracing"):
    """Экспорт PyTorch модели в ONNX"""
    print("🔄 Exporting model to ONNX...")

    model.eval()

    # Подготовить пример входных данных
    encoded = tokenizer(
        sample_input_text,
        padding=True,
        truncation=True,
        max_length=128,
        return_tensors="pt"
    )

    input_ids = encoded['input_ids']
    attention_mask = encoded['attention_mask']

    # Убедиться что модель на CPU
    model = model.cpu()

    # Экспорт в ONNX
    torch.onnx.export(
        model,
        (input_ids, attention_mask),
        save_path,
        export_params=True,
        opset_version=14,
        do_constant_folding=True,
        input_names=['input_ids', 'attention_mask'],
        output_names=['logits'],
        dynamic_axes={
            'input_ids': {0: 'batch_size', 1: 'sequence'},
            'attention_mask': {0: 'batch_size', 1: 'sequence'},
            'logits': {0: 'batch_size'}
        }
    )

    print(f"✅ ONNX model exported to: {save_path}")
    return save_path

def quantize_onnx_model(onnx_model_path, quantized_model_path):
    """Квантизация ONNX модели"""
    print("🔧 Quantizing ONNX model...")

    # Динамическая квантизация ONNX модели
    quantize_dynamic(
        onnx_model_path,
        quantized_model_path,
        weight_type=QuantType.QInt8
    )

    print(f"✅ Quantized ONNX model saved to: {quantized_model_path}")
    return quantized_model_path

def create_onnx_session(model_path, providers=None):
    """Создать ONNX Runtime сессию с CUDA (если доступна)"""
    if providers is None:
        available_providers = ort.get_available_providers()
        if 'CUDAExecutionProvider' in available_providers:
            providers = ['CUDAExecutionProvider', 'CPUExecutionProvider']
            print("⚡ Using GPU (CUDAExecutionProvider) for ONNX inference.")
        else:
            providers = ['CPUExecutionProvider']
            print("🧠 Using CPU for ONNX inference.")

    session = ort.InferenceSession(model_path, providers=providers)
    return session

def predict_with_onnx(session, input_ids, attention_mask):
    """Предсказание с ONNX моделью"""
    # Подготовить входные данные
    ort_inputs = {
        'input_ids': input_ids.numpy().astype(np.int64),
        'attention_mask': attention_mask.numpy().astype(np.int64)
    }

    # Запустить инференс
    ort_outputs = session.run(None, ort_inputs)
    logits = ort_outputs[0]

    return torch.from_numpy(logits)

def evaluate_onnx_model(session, dataloader, label_encoder, model_name="ONNX Model"):
    """Оценить производительность ONNX модели"""
    all_preds = []
    all_labels = []
    inference_times = []

    print(f"\n🔍 Evaluating {model_name}...")

    for batch in tqdm(dataloader, desc=f"Eval {model_name}"):
        input_ids = batch['input_ids']
        attention_mask = batch['attention_mask']
        labels = batch['labels']

        # Измерение времени инференса
        start_time = time.time()
        logits = predict_with_onnx(session, input_ids, attention_mask)
        end_time = time.time()
        inference_times.append(end_time - start_time)

        preds = torch.argmax(logits, dim=1)
        all_preds.extend(preds.numpy())
        all_labels.extend(labels.numpy())

    # Вычисление метрик
    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='macro')
    avg_inference_time = sum(inference_times) / len(inference_times)

    print(f"📊 {model_name} Results:")
    print(f"   Accuracy: {acc:.4f}")
    print(f"   F1 Score: {f1:.4f}")
    print(f"   Avg Inference Time: {avg_inference_time*1000:.2f} ms")

    return acc, f1, avg_inference_time, all_preds, all_labels

def compare_predictions(original_preds, quantized_preds, labels, label_encoder):
    """Сравнить предсказания оригинальной и квантизированной моделей"""
    agreement = sum(1 for o, q in zip(original_preds, quantized_preds) if o == q)
    total = len(original_preds)
    agreement_rate = agreement / total

    print(f"\n🔄 Prediction Agreement: {agreement_rate:.4f} ({agreement}/{total})")

    # Найти случаи, где модели не согласны
    disagreements = []
    for i, (o, q, true_label) in enumerate(zip(original_preds, quantized_preds, labels)):
        if o != q:
            disagreements.append({
                'index': i,
                'original': label_encoder.classes_[o],
                'quantized': label_encoder.classes_[q],
                'true': label_encoder.classes_[true_label],
                'original_correct': o == true_label,
                'quantized_correct': q == true_label
            })

    if disagreements:
        print(f"📋 First 10 disagreements:")
        for d in disagreements[:10]:
            print(f"   Sample {d['index']}: Original={d['original']} | Quantized={d['quantized']} | True={d['true']}")

    return agreement_rate, disagreements

def evaluate_pytorch_model(model, dataloader, label_encoder, device='cuda', model_name="PyTorch Model"):
    """Оценить производительность PyTorch модели"""
    model.eval()
    all_preds = []
    all_labels = []
    inference_times = []

    print(f"\n🔍 Evaluating {model_name}...")

    with torch.no_grad():
        for batch in tqdm(dataloader, desc=f"Eval {model_name}"):
            batch = {k: v.to(device) for k, v in batch.items()}

            # Измерение времени инференса
            start_time = time.time()
            outputs = model(
                input_ids=batch['input_ids'],
                attention_mask=batch['attention_mask'],
                labels=batch['labels']
            )
            end_time = time.time()
            inference_times.append(end_time - start_time)

            preds = torch.argmax(outputs["logits"], dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(batch["labels"].cpu().numpy())

    # Вычисление метрик
    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='macro')
    avg_inference_time = sum(inference_times) / len(inference_times)

    print(f"📊 {model_name} Results:")
    print(f"   Accuracy: {acc:.4f}")
    print(f"   F1 Score: {f1:.4f}")
    print(f"   Avg Inference Time: {avg_inference_time*1000:.2f} ms")

    return acc, f1, avg_inference_time, all_preds, all_labels


```

```python


def load_quantized_onnx_model():
    """Загрузить квантизированную ONNX модель"""
    kvanti_dir = "/content/drive/MyDrive/BCC/42 тыс/kvanti"

    quantized_onnx_path = os.path.join(kvanti_dir, "model_quantized.onnx")
    label_encoder_path = os.path.join(kvanti_dir, "label_encoder.pkl")
    results_path = os.path.join(kvanti_dir, "onnx_quantization_results.json")

    # Проверить наличие файлов
    if not os.path.exists(quantized_onnx_path):
        raise FileNotFoundError(f"Quantized ONNX model not found at {quantized_onnx_path}")

    if not os.path.exists(label_encoder_path):
        raise FileNotFoundError(f"Label encoder not found at {label_encoder_path}")

    # Загрузить label encoder
    with open(label_encoder_path, 'rb') as f:
        label_encoder = pickle.load(f)

    # Создать ONNX сессию
    session = create_onnx_session(quantized_onnx_path)

    # Загрузить результаты (если есть)
    results = None
    if os.path.exists(results_path):
        with open(results_path, 'r') as f:
            results = json.load(f)

    print(f"✅ Quantized ONNX model loaded from {kvanti_dir}")
    if results:
        print(f"📊 Model metrics: Acc={results['onnx_quantized']['acc']:.4f}, F1={results['onnx_quantized']['f1']:.4f}")
        print(f"💾 Size reduction: {results['size_reduction_percent']:.1f}%")
        print(f"⚡ Speedup: {results['speedup_vs_original']:.2f}x")

    return session, label_encoder, results

def predict_with_onnx_model(texts, session=None, label_encoder=None, tokenizer=None):
    """Быстрое предсказание с квантизированной ONNX моделью"""

    # Загрузить модель, если не передана
    if session is None or label_encoder is None:
        print("🔄 Loading quantized ONNX model...")
        session, label_encoder, _ = load_quantized_onnx_model()

    # Загрузить tokenizer, если не передан
    if tokenizer is None:
        from transformers import AutoTokenizer
        tokenizer = AutoTokenizer.from_pretrained("FacebookAI/xlm-roberta-base")

    predictions = []
    probabilities = []

    # Обработать тексты батчами
    batch_size = 16
    for i in tqdm(range(0, len(texts), batch_size), desc="ONNX Predicting"):
        batch_texts = texts[i:i+batch_size]

        # Токенизация
        encoded = tokenizer(
            batch_texts,
            padding=True,
            truncation=True,
            max_length=128,
            return_tensors="pt"
        )

        # Предсказание
        logits = predict_with_onnx(session, encoded['input_ids'], encoded['attention_mask'])

        # Softmax для вероятностей
        probs = torch.softmax(logits, dim=1)
        preds = torch.argmax(logits, dim=1)

        predictions.extend(preds.numpy())
        probabilities.extend(probs.numpy())

    # Преобразовать предсказания в названия классов
    predicted_labels = label_encoder.inverse_transform(predictions)

    return predicted_labels, predictions, probabilities

def demo_onnx_inference():
    """Демонстрация использования квантизированной ONNX модели"""

    sample_texts = [
        "Сделать блокировку карты.",
        "Заблокировать карту.",
        "картаны ұрлап алды, қалай бұғаттауға болады?",
        "Тревел кредиттік картасын бұғаттау."
    ]

    print("🚀 Demo: Quantized ONNX Model Inference")
    print("-" * 50)

    # Загрузить квантизированную ONNX модель
    session, label_encoder, results = load_quantized_onnx_model()

    # Предсказания
    predicted_labels, predictions, probabilities = predict_with_onnx_model(
        sample_texts, session, label_encoder
    )

    # Показать результаты
    print("\n📋 Predictions:")
    for i, (text, pred_label, prob) in enumerate(zip(sample_texts, predicted_labels, probabilities)):
        max_prob = max(prob)
        print(f"{i+1}. Text: '{text[:50]}...'")
        print(f"   Prediction: {pred_label} (confidence: {max_prob:.3f})")
        print()

    return predicted_labels, predictions, probabilities
```
